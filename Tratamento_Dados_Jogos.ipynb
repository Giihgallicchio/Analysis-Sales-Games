```python
# Importando a biblioteca pandas para manipulação de dados
import pandas as pd

# Carregando o dataset de vendas de jogos
df = pd.read_csv("Dados/vgsales.csv")

# Visualizando as primeiras linhas do dataset
df.head()
```

```python
# Exibindo informações gerais do dataset
# (tipos de dados, valores nulos e quantidade de registros)
df.info()
```

```python
# Gerando estatísticas descritivas das colunas numéricas
df.describe()
```

```python
# Verificando a quantidade de valores nulos em cada coluna
df.isnull().sum()
```

```python
# Removendo duplicados gerais
df.drop_duplicates(inplace=True)

# Preenchendo valores nulos na coluna Year com a mediana
df["Year"] = df["Year"].fillna(df["Year"].median())

# Preenchendo valores nulos na coluna Publisher
df["Publisher"] = df["Publisher"].fillna("Desconhecido")

# Convertendo o ano para inteiro
df["Year"] = df["Year"].astype(int)

# Removendo espaços extras em colunas de texto
df["Genre"] = df["Genre"].str.strip()
df["Platform"] = df["Platform"].str.strip()
```

```python
# Renomeando as colunas para português
df.columns = [
  "Ranking",
    "Nome do Jogo",
    "Plataforma",
    "Ano de Lançamento",
    "Gênero",
    "Editora",
    "Vendas América do Norte",
    "Vendas Europa",
    "Vendas Japão",
    "Outras Regiões",
    "Vendas Globais"
]

# Verificando anos inconsistentes
df["Ano de Lançamento"].sort_values().unique()

# Filtrando possíveis erros de ano
df = df[df["Ano de Lançamento"] >= 1980]
```

```python
# Criando uma coluna para verificar se a soma das regiões bate com o total global
df["Soma"] = (
    df["Vendas América do Norte"] +
    df["Vendas Europa"] +
    df["Vendas Japão"] +
    df["Outras Regiões"]
)

# Identificando possíveis inconsistências nos dados
df[abs(df["Soma"] - df["Vendas Globais"]) > 0.01]
```

```python
# Exibindo os 10 jogos mais vendidos globalmente
df.sort_values("Vendas Globais", ascending=False).head(10)
```

```python
# Verificando os jogos que mais aparecem no dataset
df["Nome do Jogo"].value_counts().head(20)
```

```python
# Listando os gêneros únicos presentes no dataset
df["Gênero"].unique()
```

```python
# Contando duplicados com base em jogo, plataforma e ano
df.duplicated(subset=["Nome do Jogo", "Plataforma", "Ano de Lançamento"]).sum()
```

```python
# Visualizando os registros duplicados para análise
duplicados = df[df.duplicated(
    subset=["Nome do Jogo", "Plataforma", "Ano de Lançamento"],
    keep=False
)]

duplicados
```

```python
# Mantendo apenas o registro com maior valor de vendas globais
df = df.sort_values("Vendas Globais", ascending=False).drop_duplicates(
    subset=["Nome do Jogo", "Plataforma", "Ano de Lançamento"]
).reset_index(drop=True)
```

```python
# Removendo coluna auxiliar utilizada na validação
df.drop(columns=["Soma"], inplace=True)
```

```python
# Conferindo se ainda existem valores nulos
df.isnull().sum()
```

```python
# Exportando o dataset tratado para Excel
df.to_excel("Dados_Tratados.xlsx", index=False)
```
